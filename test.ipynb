```python
from bs4 import BeautifulSoup
import requests

# URL of the website to be scraped
url = "https://aisaas.pkshatech.com/success/"

# Send a request to the website
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the content of the page with BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all 'a' tags within elements with class 'case-list-box'
    a_tags = soup.select('.case-list-box a')
    span_clients = soup.select('.case-lead-box span')
    # Extract the href attributes of the 'a' tags
    links = [a.get('href') for a in a_tags]
    clients = [span.text for span in span_clients]
    print(links)
    print(clients)
else:
    print("Failed to retrieve the webpage.")
```

```python
from bs4 import BeautifulSoup
import requests

# URL for the specific page to be scraped
url_iris = "https://aisaas.pkshatech.com/success/iris"

# Send a request to the website
response_iris = requests.get(url_iris)

# Check if the request was successful
if response_iris.status_code == 200:
    # Parse the content of the page with BeautifulSoup
    soup_iris = BeautifulSoup(response_iris.content, 'html.parser')

    # Define the classes to search for
    classes = [".success_in_sec2", ".success_in_sec3", ".success_in_sec4"]

    # Create a dictionary to hold the text for each class
    class_texts = {}

    for cls in classes:
        # Find the element with the specified class
        element = soup_iris.select_one(cls)

        # Extract the text from all child elements
        text = ''.join([child.get_text() for child in element.findChildren() if child.get_text().strip() != '']) if element else 'No content found'

        # Add the extracted text to the dictionary
        class_texts[cls] = text.strip()
    print(class_texts)
else:
    print("Failed to retrieve the webpage.")
```
